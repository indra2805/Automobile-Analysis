```python
#importing the libraries
 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vehicles.csv',low_memory=False)
```

```python
df = df.loc[:,['make','model','year','comb08','fuelType',]]
```

```python
df = df.loc[df.year >1999,:]
```

```python
df = df.loc[df.fuelType != ('Electricity' , 'CNG')]
```

```python
C = {'Parent_cmp' : [], 'sub_brand' : []}
cinfo = pd.DataFrame(data=C)
```

```python
cinfo
```

```python
cinfo= cinfo.append( { 'Parent_cmp' : 'Honda', 'sub_brand' : 'Honda' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Honda', 'sub_brand' : 'Acura' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Toyota', 'sub_brand' : 'Toyota' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Toyota', 'sub_brand' : 'Lexus' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Toyota', 'sub_brand' : 'Scion' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'GM', 'sub_brand' : 'General Motors' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'GM', 'sub_brand' : 'GMC' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'GM', 'sub_brand' : 'Cadillac' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Chrysler', 'sub_brand' : 'Chrysler' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Chrysler', 'sub_brand' : 'Alfa Romeo' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Chrysler', 'sub_brand' : 'Dodge' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Chrysler', 'sub_brand' : 'Fiat' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Chrysler', 'sub_brand' : 'Jeep' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Ford', 'sub_brand' : 'Ford' }, ignore_index=True  )
cinfo= cinfo.append( { 'Parent_cmp' : 'Ford', 'sub_brand' : 'Lincoln' }, ignore_index=True  )
```

```python
cinfo
```

```python
df = pd.merge(cinfo,df  , left_on = 'sub_brand' , how = 'left', right_on = 'make' , validate = 'one_to_many' )
```

```python
df = df.drop('sub_brand', axis=1)
```

```python
df = df.loc[df.fuelType != ('CNG')]
```

```python
df = df.loc[df.fuelType != ('Electricity')]
```

```python
df
```

1. For each company collect the MPG sold by each company in the years 2000-2019. Produce the box plots per company for the MPG over those years. How do the companies compare?

```python
plt.figure(figsize=(35,10))
sns.set(style="whitegrid")
sns.boxplot(data=df, x='Parent_cmp', y ='comb08')
plt.ylabel("MPG")
plt.xlabel("Company")
```

Over the years Honda has maintained a fuel effeciency well over their median fuel effeciency for their cars, and is the most fuel effecient car manufacturer. Following honda is toyota with many outliers, cars which have really good fuel effeciency. Coming in 3rd is ford woth a median fuel effeciency below 20. Followed by Chrysler and GM respectively.

2. Plot the yearly mean in the years 2000-2019 with confidence interval of the mpg for each company. That is for each company compute the mean mpg over all vehicles sold by that company per year. What changes have there been in those years? How do the companies compare?

```python
dfm = df.groupby(['Parent_cmp', 'year']).mean().reset_index()
```

```python
dfm
```

```python
plt.figure(figsize=(20,15))
sns.swarmplot(x='Parent_cmp', y='comb08', data=dfm,hue='year')
plt.ylabel("MPG")
plt.xlabel("Company")
```

All companies have a proportional increase of effeciency over the years. Honda still leading followed by toyota, ford Chrysler and GM respectively.

3. Plot the mpg for each company per year of their most fuel efficient vehicle each year. What changes have there been in those years? How do the companies compare?

```python
df2 = df.groupby(['Parent_cmp', 'year']).max().reset_index()
```

```python
df2
```

```python
plt.figure(figsize=(20,15))
sns.swarmplot(x='Parent_cmp', y='comb08', data=df2,hue='year')
plt.ylabel("MPG")
plt.xlabel("Company")
```

The most fuel effecient cars of each company has not changed since the past 3 to 5 years. Here we see toyota has the most fuel effecient car since that past four years.
